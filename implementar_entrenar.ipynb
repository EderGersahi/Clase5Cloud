## Implementar/entrenar modelo

```python
from sklearn.metrics import classification_report
from TheModel import build

import matplotlib.pyplot as plt
import numpy as np
```

```python
local_model = build.build_it()
```

```python
data = np.load('mnist_fold_eder.npz') #variar el archivo, dependiendo al nombre del integrante
x_train_fold = data['x_train_fold']
y_train_fold = data['y_train_fold']
x_val_fold = data['x_val_fold']
y_val_fold = data['y_val_fold']
```

```python
print(x_train_fold.shape)
print(x_val_fold.shape)
print(y_train_fold.shape)
print(y_val_fold.shape)
```

```python
history = local_model.fit(x_train_fold , y_train_fold, epochs=10, validation_data=(x_val_fold, y_val_fold))
y_pred = local_model.predict(x_val_fold)
y_pred_classes = np.argmax(y_pred, axis=1)
print(classification_report(y_val_fold, y_pred_classes))

plt.plot(history.history['accuracy'], label='Training accuracy')
plt.plot(history.history['val_accuracy'], label='Validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

local_model.save("model_eder.keras")#Guardar el modelo, dependiendo al nombre del integrante
```
